```python
#Calculate the total amount for each category
zone_order = df.groupby('Zone')['Orders'].sum()
state_order = df.groupby('State')['Orders'].sum().head()
product_order = df.groupby('Product_Category')['Orders'].sum().head()
product_id_order = df.groupby('Product_ID')['Orders'].sum().head()

#Create a new figure for the pie chart
plt.figure(figsize=(10, 10))

#Plotting pie chart for Zone Vs Price
plt.subplot(2, 2, 1)
plt.pie(zone_order, labels=zone_order.index, autopct='%1.f%%', colors=['lightcoral', 'lightblue', 'lightgreen', 'gold', 'lightgray'])
plt.title("Zone Vs Orders")

#Plotting pie chart for Orders Vs Price
plt.subplot(2, 2, 2)
plt.pie(state_order, labels=state_order.index, autopct='%1.f%%', colors=['lightcoral', 'lightblue', 'lightgreen', 'gold', 'lightgray'])
plt.title("Top 5 State Vs Orders")

#Plotting pie chart for Age Group Vs Price
plt.subplot(2, 2, 3)
plt.pie(product_order, labels=product_order.index, autopct='%1.f%%', colors=['lightcoral', 'lightblue', 'lightgreen', 'gold', 'lightgray'])
plt.title("Top 5 Product_Category Vs Orders")

#Plotting pie chart for Gender Vs Price
plt.subplot(2, 2, 4)
plt.pie(product_id_order, labels=product_id_order.index, autopct='%1.f%%', colors=['lightcoral', 'lightblue', 'lightgreen', 'gold', 'lightgray'])
plt.title("Top 5 Product_ID Vs Orders")

plt.tight_layout()
plt.show()
```

```python
fig, axarr = plt.subplots(4, 2, figsize=(12, 18))

df.groupby('State')['Amount'].sum().sort_values(ascending=False).plot.barh(ax=axarr[0][0], fontsize=10, color='skyblue')
axarr[0][0].set_title("Location Vs Price", fontsize=10)

df.groupby('Zone')['Amount'].sum().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=10, color='lightcoral')
axarr[0][1].set_title("Zone Vs Price", fontsize=10)

df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(10).plot.bar(ax=axarr[1][0], fontsize=8, color='lightcoral')
axarr[1][0].set_title("Product_Category Vs Price", fontsize=10)

df.groupby('Product_ID')['Amount'].sum().sort_values(ascending=False).head(10).plot.bar(ax=axarr[1][1], fontsize=10, color='lightgreen')
axarr[1][1].set_title("Product_ID Vs Price", fontsize=10)

df.groupby('Orders')['Amount'].sum().sort_values(ascending=False).plot.bar(ax=axarr[2][0], fontsize=10, color='lightgreen')
axarr[2][0].set_title("Orders Vs Price", fontsize=10)

df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False).plot.bar(ax=axarr[2][1], fontsize=10, color='gold')
axarr[2][1].set_title("Age Group Vs Price", fontsize=10)

df.groupby('Gender')['Amount'].sum().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=10, color='lightblue')
axarr[3][0].set_title("Gender Vs Orders", fontsize=10)

df.groupby('Marital_Status')['Amount'].sum().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=10, color='gold')
axarr[3][1].set_title("Car_Marital_StatusAge Vs Price", fontsize=10)

plt.subplots_adjust(hspace=0.8)
plt.subplots_adjust(wspace=.5)
sns.despine()
plt.show()
```
