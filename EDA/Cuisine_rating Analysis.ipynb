```python
cat_cols=data.select_dtypes(include=['object']).columns.tolist()
num_cols=data.select_dtypes(include=np.number).columns.tolist()
print("categorical Variables:")
print(cat_cols)
print("Numberical variable:")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    data[col].plot(kind='kde',alpha=0.6,color='red')
    plt.ticklabel_format(axis='y',style='plain')
    plt.ylabel('count')
    plt.show()
```

```python
data.columns
```

```python
cat_cols=['Location','Gender','Cuisines','Activity','Marital Status','YOB']
```

```python
for col in cat_cols:
    print(col)
    sns.countplot(x=col,data=data,order=data[col].value_counts().head(10).index,palette="Set2")
    plt.xticks(rotation=90)
    plt.show()
```

```python
corr=data[num_cols].corr()
sns.heatmap(corr,annot=True)
plt.show()
```
