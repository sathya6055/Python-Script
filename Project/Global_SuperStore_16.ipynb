# Creating New Features

```python
df['Discount']=df['Discount']*100  # converting the discount into percentage
df['Discount']=df['Discount'].astype('int') # converting the discount into integer
```

```python
df[['Sales','Quantity','Discount','Profit','Shipping Cost']].head() # checking the columns
```

Fills missing values in the 'Returned' column of the 'df' DataFrame with the string 'No'

```python
df['Returned'].fillna('No', inplace=True) # Filling missing values in 'Returned' column with 'No' to analylise the data better
```

```python
df.shape # checking the shape of the data
```

```python
df['Year']=pd.DatetimeIndex(df['Order Date']).year # extracting the year from the Order Date
df['Month']=pd.DatetimeIndex(df['Order Date']).month # extracting the month from the Order Date
```


* Year': Extracted from the 'Order Date' column, representing the year of the order.
* 'Month': Extracted from the 'Order Date' column, representing the month of the order.

```python
df['Total Sales'] = df['Sales'] * df['Quantity'] # calculating the total sales

df['Total Profit'] = df['Profit'] * df['Quantity'] # calculating the total profit

df['Total Shipping Cost'] = df['Shipping Cost'] * df['Quantity']  # calculating the total shipping cost
```

* 'Cost': Calculated as the difference between 'Sales' and 'Profit', representing the cost incurred.
* 'Total Amount': Calculated as the sum of 'Sales' and 'Shipping Cost', representing the total amount including shipping cost.

## Data Visualization


```python
cat_col = df.select_dtypes(include='object').columns.tolist() # checking the categorical columns
num_col = df.select_dtypes(include=np.number).columns.tolist() # checking the numerical columns

print(cat_col)
print(num_col)
```

```python
for col in num_col: # checking the distribution of numerical columns
    df[col].hist(bins=10, alpha=0.5,edgecolor='red',grid=False) 
    plt.ylabel('Frequency')
    plt.title(col,fontsize=12,fontweight='bold',color='green')
    plt.show()
```

* Most common shipping costs include 
* Orders are distributed across the years 2012 to 2015, with 2015 having the highest number of orders
* Orders are spread across different months, with November and December having the highest number of orders.

```python
year_analysis = df.groupby('Year').agg({ 
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum',
}).reset_index() # grouping the data by year
```

```python
year_analysis.style.background_gradient(cmap='Blues') # styling the data
```

```python
plt.figure(figsize=(12, 6)) # plotting the data

plt.subplot(1,3,1) 
plt.bar(year_analysis['Year'], year_analysis['Sales'], color=sns.color_palette('Set1'))
plt.xticks([2012,2013,2014,2015])
plt.ticklabel_format(style='plain',axis='y') # to avoid scientific notation
plt.title('Total Sales by Year')
plt.xlabel('Month')
plt.ylabel('Total Sales')

plt.subplot(1,3,2)
plt.bar(year_analysis['Year'], year_analysis['Profit'], color=sns.color_palette('Set1'))
plt.xticks([2012,2013,2014,2015])
plt.title('Total Profit by Year')
plt.xlabel('Month')
plt.ylabel('Total Profit')

plt.subplot(1,3,3)
plt.bar(year_analysis['Year'], year_analysis['Quantity'], color=sns.color_palette('Set1'))
plt.xticks([2012,2013,2014,2015])
plt.title('Total Quantity Sold by Year')
plt.xlabel('Month')
plt.ylabel('Total Quantity Sold')

plt.tight_layout() # adjusting the layout
plt.show() 
```

* Sales and profit show an increasing trend over the years, indicating positive growth in business.
* The most significant growth is observed between 2014 and 2015, with both sales and profit substantially increasing.
